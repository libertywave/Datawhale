# 处理缺失数据
### 对于数值数据，pandas使用浮点值NaN（Not a Number） 表示缺失数据，也称为哨兵值

```python
import pandas as pd
import numpy as np
string_data = pd.Series(['aardvark','artichole',np.nan,'avocado'])
string_data
```

```python
# 检测哨兵值
string_data.isnull()
```

```python
string_data[0] = None
string_data.isnull()
```

# 滤除缺失数据

```python
# Series中用dropna方法过滤缺失值
from numpy import nan as NA
data = pd.Series([1,NA,3.5,NA,7])
data.dropna()
```

```python
# 等价于
data[data.notnull()]
```

```python
# 对于DataFrame，dropna默认丢弃任何含有缺失值的行
data = pd.DataFrame([[1,6.5,3],[1.,NA,NA],[NA,NA,NA],[NA,6.5,3.]])
data
```

```python
# 传入how='all',将只丢弃全为NA的那些行
data.dropna(how='all')
```

```python
data[4] = NA
data
```

```python
# 用dropna方法丢弃全为NA的列，传入axis=1
data.dropna(how='all',axis=1)
```

```python
df = pd.DataFrame(np.random.randn(7,3))
df
```

```python
df.iloc[:4,1] = NA
df
```

```python
df.iloc[:2,2] = NA
df
```

```python
# 删除含有缺失值NA的所有行
df.dropna()
```

```python
# 传入thresh参数只留下一部分观测数据
df.dropna(thresh=2)
```

- inplace------修改调用者对象而不产生副本

```python
df
```

```python
# 调用字典，实现对不同的列填充不同的值
df.fillna({1:0.5,2:0})
```

# 数据转换

```python
# 移除重复数据
data = pd.DataFrame({'k1':['one','two'] * 3 + ['two'],
                    'k2':[1,1,2,3,3,4,4,]})
data
```

```python
# duplicated方法返回布尔型Series,表示各行是否是重复行（前面出现过得行）
data.duplicated()
```

```python
# drop_duplicates方法返回一个DataFrame,重复的shu'zu会标为Fasle
data.drop_duplicates()
```

```python
# 增加一列v1
data['v1'] = range(7)
data
```

```python
# 只根据k1列过滤重复项
data.drop_duplicates('k1')
```

# 利用函数或映射进行数据转换

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork','bacon',
                              'Pastrami', 'corned beef','Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5,6]})
data
```

```python
# 要添加一列表示该肉类食物来源的动物类型
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}
```

```python
# 转换为小写
lowercased = data['food'].str.lower()
lowercased
```

```python
# Series的map方法可以接受一个函数或者含有映射关系的字典型对象
data['animal'] = lowercased.map(meat_to_animal)
data
```

```python
# 传入能够完全完成全部工作的函数
data['food'].map(lambda x : meat_to_animal[x.lower()])
```

```python
# 替换值
data = pd.Series([1.,-999,2,-999,-1000.,3.])
data
```

```python
# -999可能表示的是缺失数据的标记值，使用replace替换为可处理的NaN值
data.replace(-999,np.nan)
```

```python
# 一次替换多个值(传入待替换值组成的列表及替换值)
data.replace([-999,-1000],np.nan)
```

```python
# 每个值有不同的替换值，传递替换列表
data.replace([-999,-1000],[np.nan,0])
```

```python
# 传入的参数也可以是字典
data.replace({-999:np.nan,-1000:0})
```

# 重命名轴索引

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                    index=['Ohio', 'Colorado', 'New York'],
                    columns=['one', 'two', 'three', 'four'])
data
```

```python
# 用map方法重命名轴
transform = lambda x:x[:4].upper()
data.index.map(transform)
```

```python
# 如果不修改原始数据，使用rename方法
data.rename(index=str.title,columns=str.upper)
```

```python
# 原地修改
data.rename(inplace=True,index={'Ohio':'vodka'})
data
```

# 离散化和面元划分
### 为了便于分析，连续数据被离散化或拆分为‘面元’

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18,25,35,60,100]
cats = pd.cut(ages,bins)
cats
```

```python
# 划分为四类
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

```python
# 传递一个列表或元组到labels，设置面元名称
group_names = ['Youth', 'YoungAdult', 'MiddleAged','Senior']
pd.cut(ages,bins,labels=group_names)
```

```python
pd.value_counts(cats)
```

# 检测和过滤异常值

```python
data = pd.DataFrame(np.random.randn(1000,4))
data
```

```python
data.describe()
```

```python
# 找出某列中绝对值大小超过3的值
col = data[2]
col[np.abs(col)>3]
```

```python
# 选出全部超过3或-3的值，使用any方法
data[(np.abs(data)>3).any(1)]
```

```python
# 将值限制在区间-3到3内
data[np.abs(data)>3] = np.sign(data)>3
data.describe()
```

```python
# 根据值是正还是负，生成1和-1
np.sign(data).head()
```
